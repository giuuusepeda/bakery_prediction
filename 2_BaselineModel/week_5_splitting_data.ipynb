## Splitting by 'Date Mask'

Write code to split your dataset into 

- a training dataset from 01.07.2013 to 31.07.2017
- a validation dataset from 01.08.2017 to 31.07.2018
- the test set from 01.08.2018 to 31.07.2019

```python
# Step 1: Convert the 'Datum' column to datetime format
df_sales["Datum"] = pd.to_datetime(df_sales["Datum"], errors="coerce", format="%Y-%m-%d")
```

```python
# Step 2: Define the masks for each split
# We use >= for the start and <= for the end to include the boundary dates
train_mask = (df_sales['Datum'] >= '2013-07-01') & (df_sales['Datum'] <= '2017-07-31')
val_mask = (df_sales['Datum'] >= '2017-08-01') & (df_sales['Datum'] <= '2018-07-31')
test_mask = (df_sales['Datum'] >= '2018-08-01') & (df_sales['Datum'] <= '2019-07-31')
```

```python
# Step 3: Apply the masks to create new DataFrames
train_df = df_sales.loc[train_mask].copy()
val_df = df_sales.loc[val_mask].copy()
test_df = df_sales.loc[test_mask].copy()
```
