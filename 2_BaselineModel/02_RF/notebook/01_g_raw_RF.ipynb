```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
df_train = pd.read_csv("/workspaces/bakery_prediction/2_BaselineModel/split_rollingaverage_data/train_data.csv", encoding = "utf-8")
df_val = pd.read_csv("/workspaces/bakery_prediction/2_BaselineModel/split_rollingaverage_data/val_data.csv", encoding = "utf-8")

df_train['Datum'] = pd.to_datetime(df_train['Datum'], format='%Y-%m-%d')
df_val['Datum'] = pd.to_datetime(df_val['Datum'], format='%Y-%m-%d')
```

```python
df_train.head()
```

```python
# alvo
TARGET = 'Umsatz'

# colunas que NÃO vão para o modelo
cols_drop = ['id', TARGET, 'Datum']

X_train = df_train.drop(columns=cols_drop)
y_train = df_train[TARGET]
X_val = df_val.drop(columns=cols_drop)
y_val = df_val[TARGET]
```

```python
#check na values y_train
print(y_train.isnull().sum())
```

```python
rf = RandomForestRegressor(
    n_estimators=300,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    n_jobs=-1,
    random_state=42
)

rf.fit(X_train, y_train)
```

```python
# previsões
y_pred_train = rf.predict(X_train)
y_pred_val = rf.predict(X_val)

def mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    # evita divisão por zero
    return np.mean(np.abs((y_true - y_pred) / np.where(y_true == 0, np.nan, y_true))) * 100

# métricas treino
mae_tr = mean_absolute_error(y_train, y_pred_train)
r2_tr = r2_score(y_train, y_pred_train)
mape_tr = mape(y_train, y_pred_train)

# métricas validação
mae_val = mean_absolute_error(y_val, y_pred_val)
r2_val = r2_score(y_val, y_pred_val)
mape_val = mape(y_val, y_pred_val)

print('--- TREINO ---')
print(f'MAE  : {mae_tr:,.2f}')
print(f'R²   : {r2_tr:,.3f}')
print(f'MAPE : {mape_tr:,.2f}%')

print('\n--- VALIDAÇÃO ---')
print(f'MAE  : {mae_val:,.2f}')
print(f'R²   : {r2_val:,.3f}')
print(f'MAPE : {mape_val:,.2f}%')


# previsões
y_pred_train = rf.predict(X_train)
y_pred_val = rf.predict(X_val)

def mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    # evita divisão por zero
    return np.mean(np.abs((y_true - y_pred) / np.where(y_true == 0, np.nan, y_true))) * 100

# métricas treino
mae_tr = mean_absolute_error(y_train, y_pred_train)
r2_tr = r2_score(y_train, y_pred_train)
mape_tr = mape(y_train, y_pred_train)

# métricas validação
mae_val = mean_absolute_error(y_val, y_pred_val)
r2_val = r2_score(y_val, y_pred_val)
mape_val = mape(y_val, y_pred_val)

print('--- TREINO ---')
print(f'MAE  : {mae_tr:,.2f}')
print(f'R²   : {r2_tr:,.3f}')
print(f'MAPE : {mape_tr:,.2f}%')

print('\n--- VALIDAÇÃO ---')
print(f'MAE  : {mae_val:,.2f}')
print(f'R²   : {r2_val:,.3f}')
print(f'MAPE : {mape_val:,.2f}%')
```
