sales by product group 2013

```python
import pandas as pd
import matplotlib.pyplot as plt

# read csv
df = pd.read_csv(r"/workspaces/bakery_prediction/0_DataPreparation/Processed/merged_data.csv")
# -------------------------------------------------------
# 1. Prepare DataFrame
# -------------------------------------------------------
df["Datum"] = pd.to_datetime(df["Datum"])

year = 2013
df_year = df[df["Datum"].dt.year == year].copy()
df_year = df_year.sort_values("Datum")

# -------------------------------------------------------
# 2. Compute 30-day rolling mean per product group
# -------------------------------------------------------
df_year["Umsatz_rolling"] = (
    df_year.groupby("Warengruppe")["Umsatz"]
           .transform(lambda x: x.rolling(window=30, min_periods=1, center=True).mean())
)

# -------------------------------------------------------
# 3. Setup: unique colors (no duplicates)
# -------------------------------------------------------
groups = df_year["Warengruppe"].unique()
colors = plt.get_cmap("tab10").colors  # 10 distinct colors

# Map each group to a color
color_map = {group: colors[i] for i, group in enumerate(groups)}

# -------------------------------------------------------
# 4. Plot
# -------------------------------------------------------
plt.figure(figsize=(14, 6))

for group in groups:
    df_g = df_year[df_year["Warengruppe"] == group]
    c = color_map[group]

    # Raw Umsatz
    plt.plot(df_g["Datum"], df_g["Umsatz"],
             color=c, alpha=0.35, linewidth=1.5,
             label=f"{group} (raw)")

    # 30-day rolling mean
    plt.plot(df_g["Datum"], df_g["Umsatz_rolling"],
             color=c, linewidth=2.5,
             label=f"{group} (30d MA)")

# -------------------------------------------------------
# 5. Styling
# -------------------------------------------------------
plt.title(f"Umsatz je Warengruppe im Jahr {year} (inkl. 30-Tage-Rolling-Average)")
plt.xlabel("Datum")
plt.ylabel("Umsatz")
plt.grid(True, alpha=0.3)

plt.legend(ncol=3)     # compact legend layout
plt.tight_layout()

plt.show()
```
